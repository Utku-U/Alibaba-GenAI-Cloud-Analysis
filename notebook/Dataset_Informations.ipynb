### **<span style="color:#8CC0EB"> Model Component Loading Latency (Service) </span>**

- **Basemodel:** If a user says "Draw a cat" or "make a sunset picture", the system uses the general cat and sunset images in its memory. If there is a WarmStart situation, the latency is close to 0. However, if the system is booting up for the first time (ColdStart), latency occurs.
- **LoRA:** These are small style patches added to the Basemodel. If the user says draw a cat but "in the style of Van Gogh" or "like the Witcher", LoRA comes into play.
- **ControlNet:** If the user provides a reference image/pose saying "make a picture with this posture", this latency comes into play.

```python
display(dataframes["basemodel_update_latency_anon.csv"].head()) # The time it takes to load the massive, heaviest model into memory for the system to run. Or the time it takes to update the basemodel loaded into the pods. 
# value: The time spent reading the Basemodel from external storage (Persistent Volume) and loading (updating) it to the GPU. 
# container_ip: The IP address of the pod where the model loading operation takes place. 

display(dataframes["lora_update_latency_anon.csv"].head()) # LoRAs are much smaller than the base model. This file contains the data of the LoRA "patch" (face recognition, style transfer, etc.) needed by the request.
# timestamp_anon: The point in time the latency was measured.
# value: The time it takes to read LoRA weights from storage areas (Persistent Volume, Object Storage) and overlay them onto the Basemodel already residing in VRAM. It should be lower than the Basemodel value.
# container_ip: The unique container_ip where the LoRA loading operation took place.

display(dataframes["controlnet_latency_data_anon.csv"].head()) # ControlNet allows us to control the structure of the generated image. ControlNet is generally larger than LoRA but smaller than the Base Model.
# timestamp_anon: The time the measurement was made. 
# value: The total time it takes to read ControlNet weights from the storage area (Volume) and load them into GPU memory (VRAM).
# container_ip: Unique IP.

dataframes["pipeline_update_latency_anon.csv"].head() # The data for the total time it takes for all components like Base Model + LoRA + ControlNet to be combined and ready to perform "inference", not just a single part.
# timestamp_anon: The time the measurement was made.
# value: Represents the loading and integration time of all components required for a request to start being processed.
# container_ip: Unique IP.
```

 ### **<span style="color:#8CC0EB"> Resource utilization monitoring data </span>**

```python
display(dataframes["pod_memory_util_anon.csv"].head()) # Shows how much RAM (system memory) each pod (container) consumes. It indicates the efficiency of the system's RAM resources. 
# value: Shows how much of the total allocated memory the pod is actively using.
# timestamp_anon: The time memory usage was measured.
# container_ip: The specific pod where the measurement was taken.

display(dataframes["pod_gpu_duty_cycle_anon.csv"].head()) # Shows at what percentage of capacity the GPU is "actively working" in a given time frame.
# value: Expresses in % (percentage) how long the GPU is actively performing calculations in a specific time frame. (100% running at full capacity, 0% pod standing by.)
# timestamp_anon: Anonymized time.
# container_ip: The specific pod.

display(dataframes["pod_gpu_memory_used_bytes_anon.csv"].head()) # Shows how much of the graphics card's own memory, which we call VRAM (Video RAM), is full in bytes. 
# timestamp_anon: Measurement time.
# value: The amount of data currently actively held in GPU memory.
# container_ip: The specific pod.
```

 ### **<span style="color:#8CC0EB"> Inference Performance Data (Output) </span>**

```python
display(dataframes["model_predict_data_anon.csv"].head()) # Pure model inference time, excluding loading time. The model is already loaded into GPU memory, all weights are ready, and only mathematical computation (matrix multiplications) is being done.
# timestamp_anon: Measurement time.
# value: The time the Generative req is actually processed on the GPU. 
# container_ip: The specific pod.

dataframes["pipeline_inference_data_anon.csv"].head() # The full response time of the inference pipeline, including model loading and inference. Covers the process from the moment the request enters the pod to the model output. (end-to-end)
# timestamp_anon: The time frame when the operation occurred.
# value: The total response time from the moment the request is accepted by the Pod.
# container_ip: The specific pod.
```

```python
dataframes["lora_request_trace.csv"].head() # Detailed inference request performance metrics, including execution time, prompt complexity, and model configuration.

# gmt_create: The exact date and time the request reached the system. 
# exec_time_seconds: The total time the request spent in the system, in days and seconds.
# predict_type: The type of operation, (TXT_2_IMG) generating images from text from scratch, or (IMG_2_IMG) image-to-image generation. 
# predict_status: Status of the operation.
# groupId: Anonymized ID of the user or organization.
# prompt_length: Length of the text.
# negative_prompt_length: E.g., bad anatomy.
# num_images_per_prompt: How many different images are requested.
# num_inference_steps: Number of denoising steps.
# checkpoint_model_version_id: Specifies which base model the request will run.
# num_lora: Specifies how many custom LoRA patches will be added to this request.
```
