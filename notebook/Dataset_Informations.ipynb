### **<span style="color:#8CC0EB"> Model Bileşen Yükleme Gecikmesi (Servis) </span>** 

- **Basemodel:** Eğer bir kullanıc "Bir kedi çiz" veya "gün batımı resmi yap" diyorsa. Sistem hafızasındaki genel kedi ve gün batımı resimlerini kullanır. Eğer bir WarmStart durumu varsa gecikme 0 a yakınıdr. Ancak sistem, ilk kez ayağa kalkıyorsa(ColdStart) gecikme yaşanır. 
- **LoRA:** Basemodel'e eklenen küçük stil yamalarıdır. Eğer kullanıcı, bir kedi çiz ama "Van Gogh tarzında" olsun veya "Witcher" gibi olsun diyorsa LoRA devreye girer.
- **ControlNet:** Eğer kullanıcı "şu duruştaki gibi bir resim yap" diyerek referans görsel/poz veriyorsa gecikme devreye girer.

```python
display(dataframes["basemodel_update_latency_anon.csv"].head()) # Sistemin çalışması için en büyük ve en ağır olan devasa modelin belleğe yüklenmesi süresi. Ya da podlara yüklenmiş olan basemodel'in güncellenmesi süresi. 
# value: Basemodel'in dış dikten(Persistent Volume) okunup GPU'ya yüklenmesinde(güncellenmesinde) geçen süre. 
# container_ip: Model yükleme işleminin gerçekleştiği pod'un IP adresi. 

display(dataframes["lora_update_latency_anon.csv"].head()) # LoRA'lar temel modele göre çok daha küçüktür. Bu dosya, isteğin ihtiyaç duyduğu LoRA "yamasının"(yüz tanıma, stil aktarımı vb.) verisidir.
# timestampt_anon: Gecikmenin ölçüldüğü zaman noktası.
# value: LoRA ağırlıklarının depolama alanlarından(Persistent Volume, Object Storage) okunup, hali hazırda VRAM de duran Basemodel üzerine bindirilmesi için geçen süre. Basemodel değerinden daha düşük olmalıdır.
# container_ip: LoRA yükleme işlenin gerçekleştiği benzersiz container_ip

display(dataframes["controlnet_latency_data_anon.csv"].head()) # ControlNet, üretilen görselin yapısını kontrol etmemizi sağlar. ControlNet genellikle LoRA'dan daha büyük, temel modelden (Base Model) ise daha küçüktür.
# timestampt_anon: ölçümün yapıldığı zaman. 
# value: ControlNet ağırlıklarının depolama alanından(Volume) okunup GPU belleğine (VRAM) yüklenmesi için geçen toplam süre.
# container_ip: benzersiz IP

dataframes["pipeline_update_latency_anon.csv"].head() # Sadece bir parçanın değil; Temel Model + LoRA + ControlNet gibi tüm bileşenlerin bir araya getirilip "çıkarım (inference)" yapmaya hazır hale gelmesi için geçen toplam süredir verisi.
# timestampt_anon: Ölçümün yapıldığı zaman.
# value: Bir isteğin, işlenmeye başlaması için gereken tüm bileşenlerin, yüklenme ve birbirine entegre edilme süresini ifade eder.
# container_ip: benzrsiz IP.
```
